```python
from tkinter import *
from tkinter import messagebox
import random
import tempfile
import webbrowser
import urllib.parse
from datetime import datetime, timedelta
from calendar import monthrange
from tkinter import ttk
from tkcalendar import DateEntry

# calculation
def total():
    # ticket prices
    global Economy_price, Premium_price, Business_price, First_class_price
    
    Economy_price = int(Economy_entry.get()) * 5000
    Premium_price = int(Premium_entry.get()) * 10000
    Business_price = int(Business_entry.get()) * 15000
    First_class_price = int(First_class_entry.get()) * 20000
    
    total_ticket = Economy_price + Premium_price + Business_price + First_class_price
    costofticket_entry.delete(0, END)
    costofticket_entry.insert(0, total_ticket + .00)
    
    # taxes
    global tax_total
    tax_total = total_ticket * 18 / 100
    tax_entry.delete(0, END)
    tax_entry.insert(0, tax_total)
    
    # grand total
    global grand_total
    grand_total = total_ticket + tax_total
    totalbutton_entry.delete(0, END)
    totalbutton_entry.insert(0, grand_total + .00)
    
    # update the textarea with the grand total value
    textarea.delete(1.0, END)
    
def bill_area():
    total()  # Call the total function to update the grand_total variable
    name = nameEntry.get().strip()
    phone = phoneEntry.get().strip()
    ticket = costofticket_entry.get().strip()
    tax = tax_entry.get().strip()
    departing = departing_entry.get().strip()
    arriving = arriving_entry.get().strip()
    flight_name = flight_name_var.get().strip()
    ticket_id = ticket_id_entry.get().strip()
    departing_date = departing_date_entry.get().strip()
    departing_time = departing_time_entry.get().strip()
    
    if not name or not phone:
        messagebox.showerror("Error", "Customer details are required")
    elif not ticket and not tax:
        messagebox.showerror("Error", "Ticket details are required")
    elif not departing or not arriving:
        messagebox.showerror("Error", "Departing and arriving places are required")
    elif not flight_name or not ticket_id:
        messagebox.showerror("Error", "Flight name and ticket ID are required")
    else:
        textarea.delete(1.0, END)
        textarea.tag_configure("center", justify='center')
        textarea.tag_configure("large_font", font=("DejaVu Sans Mono", 10))
        textarea.tag_configure("bold_font", font=("DejaVu Sans Mono", 11, "bold"))
        textarea.insert(END, "WELCOME TO OUR AIRPLANE TICKET BOOKING SYSTEM\n", "center")
        textarea.insert(END, f"\nDate & Time\t\t : \t{now}\n", "large_font")
        textarea.insert(END, f"Booking number\t\t : \t{bookingnumber}\n", "large_font")
        textarea.insert(END, f"Customer\t\t : \t{nameEntry.get()}\n", "large_font")
        textarea.insert(END, f"Phone number\t\t : \t{phoneEntry.get()}\n", "large_font")
        textarea.insert(END, f"Flight Name\t\t : \t{flight_name_entry.get()}\n", "large_font")
        textarea.insert(END, f"Ticket ID\t\t : \t{ticket_id_entry.get()}\n", "large_font")
        textarea.insert(END, f"Departing from\t\t : \t{departing_entry.get()} on {departing_date_entry.get()} at {departing_time_entry.get()}\n", "large_font")
        textarea.insert(END, f"Arriving at\t\t : \t{arriving_entry.get()}\n", "large_font")
        textarea.insert(END, f"==========================================================")
        textarea.insert(END, "\nFLIGHT DETAILS\t\t\tPRICE", "large_font")
        textarea.insert(END, f"\n==========================================================")
        if Economy_entry.get() != "0":
            textarea.insert(END, f"\t\nEconomy\t\t\t{Economy_entry.get()}\t\t\t{Economy_price}.00", "large_font")
        if Premium_entry.get() != "0":
            textarea.insert(END, f"\t\nPremium\t\t\t{Premium_entry.get()}\t\t\t{Premium_price}.00", "large_font")
        if Business_entry.get() != "0":
            textarea.insert(END, f"\t\nBusiness\t\t\t{Business_entry.get()}\t\t\t{Business_price}.00", "large_font")
        if First_class_entry.get() != "0":
            textarea.insert(END, f"\t\nFirst Class\t\t\t{First_class_entry.get()}\t\t\t{First_class_price}.00", "large_font")
        textarea.insert(END, f"\n----------------------------------------------------------")
        if tax_entry.get():
            textarea.insert(END, f"\t\nCGST @ 9%\t\t\t\t\t\t{tax_total/2}", "large_font")
            textarea.insert(END, f"\t\nSGST @ 9%\t\t\t\t\t\t{tax_total/2}", "large_font")
            textarea.insert(END, f"\n**********************************************************\n")
            textarea.insert(END, f"NET AMOUNT\t\t\t\t\t\t{grand_total:.2f}\n", "large_font")
            textarea.insert(END, "Thank you for booking with us\n", "center")
            textarea.insert(END, "'Safe travels'\n", "center")
            x = ticket_id_entry.insert(0, ''.join(random.choice('ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789') for _ in range(10)))
            textarea.insert(END, x, "large_font")

def save_bill():
    global bookingnumber
    result = messagebox.askyesno("Confirm", "Do you want to save the bill")
    if result:
        bill_content = textarea.get(1.0, END)
        file = open(f"bills/{bookingnumber}.txt", "w")
        file.write(bill_content)
        file.close()
        messagebox.showinfo("Success", f"{bookingnumber} is saved Successfully")


def print_bill():
    if textarea.get(1.0, END) == "\n":
        messagebox.showerror("Error", "Your bill is empty")
    else:
        file = tempfile.mktemp(".txt")
        open(file, "w").write(textarea.get(1.0, END))
        os.startfile(file)

def share_on_whatsapp():
    phone_number = phoneEntry.get().strip()
    if not phone_number:
        messagebox.showerror("Error", "Phone number is required")
        return
    # Get the content of the bill
    bill_content = textarea.get(1.0, END).strip()
    if not bill_content:
        messagebox.showerror("Error", "Bill content is empty")
        return
    # Encode the bill content
    encoded_bill_content = urllib.parse.quote(bill_content)
    # Create the WhatsApp URL
    whatsapp_url = f"https://wa.me/{phone_number}?text={encoded_bill_content}"
    # Open the WhatsApp URL
    webbrowser.open(whatsapp_url)

def clear_all():
    Economy_entry.delete (0, END)
    Premium_entry.delete(0, END)
    Business_entry.delete(0, END)
    First_class_entry.delete(0, END)
    
    Economy_entry.insert(0 , 0)
    Premium_entry.insert(0, 0)
    Business_entry.insert(0, 0)
    First_class_entry.insert(0, 0)
    
    costofticket_entry.delete(0, END )
    tax_entry.delete(0, END)
    totalbutton_entry.delete(0, END)
    
    nameEntry.delete(0, END)
    phoneEntry.delete(0, END)
    departing_entry.delete(0, END)
    arriving_entry.delete(0, END)
    flight_name_entry.delete(0, END)
    ticket_id_entry.delete(0, END)
    departing_date_entry.delete(0, END)
    departing_time_entry.delete(0, END)
    textarea.delete(1.0, END)

# root
bookingnumber = random.randint(300001, 800000)
root = Tk()
root.title("Airplane Ticket Booking System")
# root.iconbitmap("bill.ico")
now = datetime.now()
# Maximize window size
root.state('zoomed')

# title
head = Label(root, text="AIRPLANE TICKET BOOKING SYSTEM", font=("ARIAL BLACK", 25), bg="#32CD32", fg="white", bd=8, relief=GROOVE)
head.pack(fill=X)

# billing area
items4 = Frame(root)
items4.pack(side=RIGHT)
billing = Frame(items4, bd=8, relief=GROOVE)
billing.pack(side=TOP, fill=BOTH, expand=True)
# root
bill = Label(billing, text="BILLING AREA", font=("Segoe UI", 15, "bold"), bd=8, relief=GROOVE)
bill.pack(side=TOP, fill=X)

# scrollbar for bill area
scroll = Scrollbar(billing, orient=VERTICAL)  # scrollbar
scroll.pack(side=RIGHT, fill=Y)
textarea = Text(billing, height=31, width=58, yscrollcommand=scroll.set)
textarea.pack(side=LEFT, fill=BOTH, expand=True)
scroll.config(command=textarea.yview)

# customer info
customer = LabelFrame(root, text="CUSTOMER DETAILS", font=("Segoe UI", 15, "bold"), fg="black", bd=8, relief=SUNKEN, bg="#C6F4D6")
customer.pack(fill=X)

name = Label(customer, text="NAME", font=("Segoe UI", 15, "bold"), bg="#C6F4D6")
name.grid(row=1, column=0, padx=10, pady=10)
nameEntry = Entry(customer, font=("ARIAL", 15, "bold"), bd=7, width=18)
nameEntry.grid(row=1, column=1, padx=25)

phone = Label(customer, text="PHONE NUMBER", font=("Segoe UI", 15, "bold"), bg="#C6F4D6")
phone.grid (row=1, column=2, padx=10, pady=10)
phoneEntry = Entry(customer, font=("ARIAL", 15, "bold"), bd=7, width=18)
phoneEntry.grid(row=1, column=3, padx=25)

# flight info
flight = LabelFrame(root, text="FLIGHT DETAILS", font=("Segoe UI", 15, "bold"), fg="black", bd=8, relief=SUNKEN, bg="#C6F4D6")
flight.pack(fill=X)

flight_name = Label(flight, text="FLIGHT NAME", font=("Segoe UI", 15, "bold"), bg="#C6F4D6")
flight_name.grid(row=1, column=0, padx=10, pady=10)
flight_name_var = StringVar()
flight_name_entry = ttk.Combobox(flight, textvariable=flight_name_var, font=("ARIAL", 15, "bold"), width=18)
flight_name_entry['values'] = ['British Airways', 'Emirates', 'American Airlines', 'United Airlines', 'Southwest Airlines', 'Cathay Pacific', 'Middle East Airlines', 'Air India', 'IndiGo', 'SpiceJet', 'Alliance Air', 'Air France', 'Austrian Airlines', 'Singapore Airlines', 'Delta Air Lines']
flight_name_entry.grid(row=1, column=1, padx=25)

ticket_id = Label(flight, text="TICKET ID", font=("Segoe UI", 15, "bold"), bg="#C6F4D6")
ticket_id.grid(row=1, column=2, padx=10, pady=10)
ticket_id_entry = Entry(flight, font=("ARIAL", 15, "bold"), bd=7, width=18)
ticket_id_entry.insert(0, ''.join(random.choice('ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789') for _ in range(10)))
ticket_id_entry.grid(row=1, column=3, padx=25)

departing = Label(flight, text="DEPARTING FROM", font=("Segoe UI", 15, "bold"), bg="#C6F4D6")
departing.grid(row=2, column=0, padx=10, pady=10)
departing_entry = ttk.Combobox(flight, font=("ARIAL", 15, "bold"), width=18)
departing_entry['values'] = ['India', 'China', 'America', 'Indonesia', 'Brazil', 'Pakistan', 'Nigeria', 'Bangladesh', 'Russia', 'Mexico', 'Japan', 'Germany', 'Philippines', 'Vietnam', 'Ethiopia', 'Egypt', 'Iran', 'France', 'Spain']
departing_entry.grid(row=2, column=1, padx=25)

departing_date = Label(flight, text="DEPARTING DATE", font=("Segoe UI", 15, "bold"), bg="#C6F4D6")
departing_date.grid(row=2, column=2, padx=10, pady=10)
departing_date_entry = DateEntry(flight, font=("ARIAL", 15, "bold"), width=18)
departing_date_entry.grid(row=2, column=3, padx=25)

departing_time = Label(flight, text="DEPARTING TIME", font=("Segoe UI", 15, "bold"), bg="#C6F4D6")
departing_time.grid(row=3, column=0, padx=10, pady=10)
departing_time_entry = Spinbox(flight, from_=0, to=23, width=18, font=("ARIAL", 15, "bold"))
departing_time_entry.grid(row=3, column=1, padx=25)

arriving = Label(flight, text="ARRIVING AT", font=("Segoe UI", 15, "bold"), bg="#C6F4D6")
arriving.grid(row=3, column=2, padx=10, pady=10)
arriving_entry = ttk.Combobox(flight, font=("ARIAL", 15, "bold"), width=18)
arriving_entry['values'] = ['India', 'China', 'America', 'Indonesia ', 'Brazil', 'Pakistan', 'Nigeria', 'Bangladesh', 'Russia', 'Mexico', 'Japan', 'Germany', 'Philippines', 'Vietnam', 'Ethiopia', 'Egypt ', 'Iran', 'France', 'Spain']
arriving_entry.grid(row = 3, column=3, padx=25)

# ticket info
items = Frame(root)
items.pack(fill=X)

items1 = LabelFrame(items, text="TICKET DETAILS", font=("Segoe UI", 18, "bold"), fg="black", bg="#C6F4D6")
items1.grid(row=0, column=1, padx=10)

# ticket items
Economy = Label(items1, text="Economy", font=("Segoe UI", 15, "bold"), bg="#C6F4D6")
Economy.grid(row=1, column=0, padx=30)
Economy_entry = Entry(items1, font=("ARIAL",  15, "bold"), width=10, bd=5)
Economy_entry.grid(row=1, column=1, padx=40, pady=5)
Economy_entry.insert(0, 0)

Premium = Label(items1, text="Premium", font=("Segoe UI", 15, "bold"), bg="#C6F4D6")
Premium.grid(row=2, column=0, padx=30)
Premium_entry = Entry(items1, font=("ARIAL", 15, "bold"), width=10, bd=5)
Premium_entry.grid(row=2, column=1, padx=40, pady=5)
Premium_entry.insert(0, 0)

Business = Label(items1, text="Business", font=("Segoe UI", 15, "bold"), bg="#C6F4D6")
Business.grid(row=3, column=0, padx=30)
Business_entry = Entry(items1, font=("ARIAL", 15, "bold"), width=10, bd=5)
Business_entry.grid(row=3, column=1, padx=40, pady=5)
Business_entry.insert(0, 0)

First_class = Label(items1, text="First Class", font=("Segoe UI", 15, "bold"), bg="#C6F4D6")
First_class.grid(row=4, column=0, padx=30)
First_class_entry = Entry(items1, font=("ARIAL", 15, "bold"), width=10, bd=5)
First_class_entry.grid(row=4, column=1, padx=40, pady=10)
First_class_entry.insert(0, 0)

# billing section
billsect = LabelFrame(root, font=("Segoe UI", 15, "bold"), fg="black", bg="#C6F4D6")
billsect.pack(fill=X)

costofticket = Label(billsect, text="COST OF TICKET", font=("Segoe UI", 11, "bold"), bg="#C6F4D6")
costofticket.grid(row=0, column=0, padx=5)
costofticket_entry = Entry(billsect, font=("ARIAL",  17, "bold"), width=12,bd=4)
costofticket_entry.grid(row=0, column=1, padx=5, pady=10)

tax = Label(billsect, text="TAX", font=("Segoe UI", 11, "bold"), bg="#C6F4D6")
tax.grid(row=1, column=0, padx=5)
tax_entry = Entry(billsect, font=("ARIAL", 17, "bold"), width=12, bd=4)
tax_entry.grid(row=1, column=1, padx=5, pady=10)

# total
billsect2 = LabelFrame(root, font=("Segoe UI", 15, "bold"), fg="black", bg="#C6F4D6")
billsect2.pack(fill=X)

totalbutton = Button(billsect2, text="TOTAL", font=("Segoe UI", 12, "bold"), bd=5, bg="#32CD32", fg="white", width=17, relief=GROOVE, command=total)
totalbutton.grid(row=0, column=1, padx=5)
grandtotal = Label(billsect2, text="GRAND TOTAL", font=("Segoe UI", 15, "bold"), bg="#C6F4D6")
grandtotal.grid(row=0, column=2, padx=5)
totalbutton_entry = Entry(billsect2, font=("ARIAL", 20, "bold"), width=22, bd=5)
totalbutton_entry.grid(row=0, column=3, padx=5, pady=15)

# button
buttonframe = LabelFrame(items4, font=("Segoe UI", 15, "bold"), fg="black", bg="#C6F4D6", bd=5)
buttonframe.pack(fill=X)

button_bill = Button(buttonframe, text="BILL", font=("Segoe UI", 12, "bold"), bd=5, bg="#32CD32", fg="white", width=8, relief=GROOVE, command=bill_area)
button_bill.grid(row=0, column=1, padx=10, pady=10)

button_print = Button(buttonframe, text="PRINT", font=("Segoe UI", 12, "bold"), bd=5, bg="#32CD32", fg="white", width=8, relief=GROOVE, command=print_bill)
button_print.grid(row=0, column=2, padx=10, pady=10)

button_share = Button(buttonframe, text="SHARE ON WA", font=("Segoe UI", 12, "bold"), bd=5, bg="#32CD32", fg="white", width=12, relief=GROOVE, command=share_on_whatsapp)
button_share.grid(row=0, column=3, padx=10, pady=10)

button_clear = Button(buttonframe, text="CLEAR", font=("Segoe UI", 12, "bold"), bd=5, bg="#32CD32", fg="white", width=8, relief=GROOVE, command=clear_all)
button_clear.grid(row=0, column=4, padx=10, pady=10)

footer = Label(root, text=" 2024 Airplane Ticket Booking System | Contact us at info@airplaneticketbooking.com", bg="#C6F4D6", fg="black", font=("Segoe UI", 12, "bold"), pady=10, relief=GROOVE, bd=2)
footer .pack(side=BOTTOM, fill=X)

root.mainloop()
```
